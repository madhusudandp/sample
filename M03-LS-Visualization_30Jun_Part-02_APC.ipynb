### Date handling on Superstore data

```python
df_superstore.info()
```

#### Change Order and Ship date to `date time format`

`pd.to_datetime()`

```python
df_superstore['Order Date'] = pd.to_datetime(df_superstore['Order Date'])
df_superstore['Ship Date'] = pd.to_datetime(df_superstore['Ship Date'])
```

```python
df_superstore.info()
```

```python
df_superstore.head(2).T
```

#### Extract month and year

`Datetime.strftime()`

```python
import datetime
```

```python
df_superstore['Order_Year'] = df_superstore['Order Date'].dt.strftime('%Y')
df_superstore['Order_Month'] = df_superstore['Order Date'].dt.strftime('%m')
```

```python
df_superstore.head(2).T
```

#### Create 'Order_Year_Month' Column : `2017-11`

```python
df_superstore['Order_Year_Month'] = df_superstore['Order Date'].dt.to_period('M') #remember
```

```python
df_superstore.sample(2).T
```

#### Concatenate

```python
df_superstore['Order_Year_Month_concat'] = df_superstore['Order_Year'] + "-"+df_superstore['Order_Month']
```

```python
df_superstore.sample(2).T
```

### Superstore sales data for how many years

```python
df_superstore['Order_Year'].unique()
```

`4 years of data: 2015 to 2018`

####  subset the data for the last two years

```python
filt_1 = df_superstore['Order_Year']>'2016'
```

```python
df_superstore[filt_1]['Order_Year'].unique()
```

```python
df_superstore_sub = df_superstore[filt_1]
```

#### let us the plot the line chart

```python
df_superstore.info()
```

```python
plt.subplots(figsize=(18,6))
plt.plot(df_superstore_sub['Order_Year_Month_concat'], df_superstore_sub['Sales'])
plt.show()
```

### **Observation: 1 Data is not sorted**

```python
df_superstore_sub_sorted = df_superstore_sub.sort_values(by='Order_Year_Month_concat')
```

```python
plt.subplots(figsize=(18,6))
plt.plot(df_superstore_sub_sorted['Order_Year_Month_concat'], df_superstore_sub['Sales'])
plt.show()
```

### **Observation: 2 Data is sorted but not aggregated**

```python
df_superstore_sub_sorted_agg = df_superstore_sub_sorted.groupby('Order_Year_Month_concat').agg(Total_Sales = ('Sales', 'sum'), Total_Profit = ('Profit', 'sum')).reset_index()
```

```python
df_superstore_sub_sorted_agg
```

```python
plt.subplots(figsize=(18,6))
plt.plot(df_superstore_sub_sorted_agg['Order_Year_Month_concat'], df_superstore_sub_sorted_agg['Total_Sales'])
plt.show()
```

```python
plt.subplots(figsize=(16,6))
plt.plot(df_superstore_sub_sorted_agg['Order_Year_Month_concat'], df_superstore_sub_sorted_agg['Total_Sales'], color='purple', marker='o', linestyle='-', label='Total Sales')  #total sales    
plt.plot(df_superstore_sub_sorted_agg['Order_Year_Month_concat'], df_superstore_sub_sorted_agg['Total_Profit'], color='green', marker='o', linestyle='--', label='Total Profit')  #total profit    
plt.title('Month over Month Total Sales')
plt.xlabel('Order Year Month')
plt.legend()
plt.show()
```

`| '-' | '--' | '-.' |`

#### profit as an area chart

```python
plt.subplots(figsize=(16,6))
plt.plot(df_superstore_sub_sorted_agg['Order_Year_Month_concat'], df_superstore_sub_sorted_agg['Total_Sales'], color='purple', marker='o', linestyle='-', label='Total Sales')  #total sales    
plt.fill_between(df_superstore_sub_sorted_agg['Order_Year_Month_concat'], df_superstore_sub_sorted_agg['Total_Profit'], color='green', linestyle='--', label='Total Profit')  #total profit    
plt.title('Month over Month Total Sales')
plt.xlabel('Order Year Month')
plt.legend()
plt.show()
```
